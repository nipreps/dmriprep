```python
import os
import json
from glob import glob

from collections import defaultdict

import nibabel as nib
```

```python
dwi_jsons = sorted(glob('/archive/data/SPINS/data/nii/SPN01_*_0*/*DTI*json'))

slice_timings = []

for i in dwi_jsons:
    with open(i, 'r') as f:
        json_dict = json.load(f)
    try:
        timing = json_dict['SliceTiming']
        if not timing in slice_timings:
            slice_timings.append(timing)
    except:
        print(i)
        continue
```

```python
dwi_jsons = sorted(glob('/archive/data/SPINS/data/nii/SPN01_*_0*/*DTI*json'))

slice_orders = []

for i in dwi_jsons:
    with open(i, 'r') as f:
        json_dict = json.load(f)
    try:
        timing = json_dict['SliceTiming']
        slice_order = sorted(range(len(timing)),key=timing.__getitem__)
        if not slice_order in slice_orders:
            slice_orders.append(slice_order)
    except:
        continue
```

```python
dwi_niis = sorted(glob('/archive/data/SPINS/data/nii/SPN01_*_0*/*DTI*nii.gz'))

unique_shapes = defaultdict(set)

for i in dwi_niis:
    site = i.split('/')[6].split('_')[1]
    data = nib.load(i)
    if not data.shape[2] == 66:
        print(i)
    unique_shapes[site].add(data.shape)
```

```python
dwi_niis = sorted(glob('/archive/data/SPINS/data/nii/SPN01_*_0*/*DTI*nii.gz'))

for i in dwi_niis:
    data = nib.load(i)
    slice_num = data.shape[2]
    slices = list(range(1, slice_num))
    even_slices = [slice for slice in slices if slice % 2 == 0]
    odd_slices = [slice for slice in slices if slice % 2 == 1]
    slice_reorder = odd_slices + even_slices
    
```
